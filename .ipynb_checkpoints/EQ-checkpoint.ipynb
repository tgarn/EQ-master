```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import train_test_split

from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, mean_squared_error, precision_recall_fscore_support, accuracy_score
```

```python
eq = pd.read_csv("EQ.csv")
img=mpimg.imread('map.png')
imgplot = plt.imshow(img)
```

```python
plt.scatter(eq.Longitude, eq.Latitude,color='b')
plt.xlabel("Actual y Value")
plt.ylabel("Predicted y Value")

plt.show()
print('A map of all latitude and longitude, if you look at the location of the points you can make out locations in the world such as the Affrican coast')
```

```python
X = eq[["Latitude","Longitude","Depth"]]
y = eq["Magnitude"]

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, train_size=0.5, random_state=0)

lnmod = LinearRegression()
lnmod.fit(X_train1,y_train1)
pred1 = lnmod.predict(X_test1)
mse1 = mean_squared_error(y_test1, pred1)
print("Linear Regression MSE for predictiong Magnitude: ",mse1)
```

```python
kf1 = KFold(n_splits=10)
kf2 = KFold(n_splits=50)
kf3 = KFold(n_splits=100)
#using 3 diffrent splits 10,50 and 100 splits
preds_cv1 = []
preds_cv2 = []
preds_cv3 = []
for train, test in kf1.split(X):
    X_cv1 = X.iloc[train]
    y_cv1 = y.iloc[train]
    X_cv_test1 = X.iloc[test]
    lm_cv1 = Lasso(max_iter=5000, tol=0.001)
    lm_cv1.fit(X_cv1, y_cv1)
    preds1 = lm_cv1.predict(X_cv_test1)
    preds_cv1 += list(preds1)




for train, test in kf2.split(X):
    X_cv2 = X.iloc[train]
    y_cv2 = y.iloc[train]
    X_cv_test2 = X.iloc[test]
    lm_cv2 = Lasso(max_iter=5000, tol=0.001)
    lm_cv2.fit(X_cv2, y_cv2)
    preds2 = lm_cv2.predict(X_cv_test2)
    preds_cv2 += list(preds2)


for train, test in kf3.split(X):
    X_cv3 = X.iloc[train]
    y_cv3 = y.iloc[train]
    X_cv_test3 = X.iloc[test]
    lm_cv3 = Lasso(max_iter=5000, tol=0.001)
    lm_cv3.fit(X_cv3, y_cv3)
    preds3 = lm_cv3.predict(X_cv_test3)
    preds_cv3 += list(preds3)
    
mse_cv1 = mean_squared_error(y, preds_cv1)
mse_cv2 = mean_squared_error(y, preds_cv2)
mse_cv3 = mean_squared_error(y, preds_cv3)
print ("MSE 10 split cross Validation:", mse_cv1, "  50 split:", mse_cv2,
       "  100 split:", mse_cv3)
```

```python
#since the data set is massive decided to try using a higher number of splits 
#Using the laso the mse is lowered giving better results then just using linear regression
```

```python
X = eq[["Latitude","Longitude"]]
y = eq["Magnitude"]
kf1 = KFold(n_splits=10)
kf2 = KFold(n_splits=50)
kf3 = KFold(n_splits=100)
#doing the same thing as above with the exeption of excluding depth as a factor
preds_cv1 = []
preds_cv2 = []
preds_cv3 = []

for train, test in kf1.split(X):
    X_cv1 = X.iloc[train]
    y_cv1 = y.iloc[train]
    X_cv_test1 = X.iloc[test]
    lm_cv1 = Lasso(max_iter=5000, tol=0.001)
    lm_cv1.fit(X_cv1, y_cv1)
    preds1 = lm_cv1.predict(X_cv_test1)
    preds_cv1 += list(preds1)

for train, test in kf2.split(X):
    X_cv2 = X.iloc[train]
    y_cv2 = y.iloc[train]
    X_cv_test2 = X.iloc[test]
    lm_cv2 = Lasso(max_iter=5000, tol=0.001)
    lm_cv2.fit(X_cv2, y_cv2)
    preds2 = lm_cv2.predict(X_cv_test2)
    preds_cv2 += list(preds2)

for train, test in kf3.split(X):
    X_cv3 = X.iloc[train]
    y_cv3 = y.iloc[train]    
    X_cv_test3 = X.iloc[test]
    lm_cv3 = Lasso(max_iter=5000, tol=0.001)
    lm_cv3.fit(X_cv3, y_cv3)
    preds3 = lm_cv3.predict(X_cv_test3)
    preds_cv3 += list(preds3)
    
mse_cv1 = mean_squared_error(y, preds_cv1)
mse_cv2 = mean_squared_error(y, preds_cv2)
mse_cv3 = mean_squared_error(y, preds_cv3)
print ("MSE 10 split cross Validation:", mse_cv1, "  50 split:", mse_cv2,
       "  100 split:", mse_cv3)
```

```python
#removing depth as a factor had very little impact on the overall mse even having lower mse with 10 splits without depth
```
